### IMPORTs

```python
from dotenv import load_dotenv, find_dotenv
import pandas as pd
import os
```

### Carregando dados de Lojas

```python
# Ele procura o .env sozinho, não importa em qual subpasta o notebook esteja!
load_dotenv(find_dotenv())

# Puxa a variável direto do sistema
diretorio_csv = os.getenv('CAMINHO_DIRETORIO_CSV')

caminho_lojas_csv = os.path.join(diretorio_csv, "lojas_vendedores.csv")
df_lojas = pd.read_csv(caminho_lojas_csv)

display(df_lojas)
```

### Criando PARQUET

```python
# O find_dotenv() "caça" o arquivo .env automaticamente nas pastas acima
load_dotenv(find_dotenv())

# 1. Puxa o caminho da pasta salvo no arquivo .env
diretorio_base = os.getenv('CAMINHO_DIRETORIO_PARQUET')

# 2. Cria o caminho completo (Pasta + Nome do Arquivo)
caminho_completo = os.path.join(diretorio_base, "lojas_vendedores_prata.parquet")

# 3. Salva o DataFrame no diretório especificado
df_lojas.to_parquet(caminho_completo, engine="pyarrow", index=False)

# Print seguro: confirma a criação sem expor o caminho 'E:\...'
print("✅ Arquivo 'lojas_vendedores_prata.parquet' criado com sucesso no diretório configurado!")
```
